```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_alphas = 10
alphlowlim = 0.0
alphhighlim = 0.9
num_ds = 50
dlowlim = 0.0
dhighlim = 5.0

s_vals = np.linspace(1.0, 5.0, 101)
s_vals = s_vals[:-1]
```

```python
def contour_plt(P_alpha_d, i=0, save=False, name=""):
    """
    Visualise P_alpha_d as a countour plot using contourf
    """
    # plot
    # rescale the y axis to be in the range of 0 to 2
    # rescale the x axis to be in the range of 0 to 1000
    plt.figure(figsize=(4, 4))
    x = np.linspace(dlowlim, dhighlim, num_ds+1)
    x = x[:-1]
    y = np.linspace(alphlowlim, alphhighlim, num_alphas)
    plt.contourf(x, y, P_alpha_d[i], 20, cmap='RdGy')
    plt.xlabel('d')
    plt.ylabel('alpha')
    plt.colorbar()
    if save:
        plt.savefig(name)
        plt.close()
    else:
        plt.show()
```

```python
P_alpha_d = np.load("P_alpha_d_run3.npy")
```

```python
# for i in range(0, len(P_alpha_d), s_vals.shape[0]):
for i in range(0, len(P_alpha_d)):
    contour_plt(P_alpha_d, i)
```
