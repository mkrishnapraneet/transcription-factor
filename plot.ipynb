```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot data from file for MSD (fort.9)

# load data from file
data = np.loadtxt('fort.9')

# shape of the data
print(data.shape)
# print(data)

# plot the data
plt.plot(data[:,0], data[:,1])
plt.title("Mean Square Displacement")
plt.xlabel("Lag Time (frames)")
plt.ylabel("MSD (pixels^2)")
```
